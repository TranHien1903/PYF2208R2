```python

from tkinter import *
from tkinter import ttk
from tkinter import messagebox
class Staff:
    def __init__(self, id, name, sex, division, phone, email, address):
        self.id = id
        self.name = name
        self.sex = sex
        self.division = division
        self.phone = phone
        self.email = email
        self.address = address
list_staff = []
def add_info():
    list_staff.append(Staff(entry_id.get(), entry_name.get(), combo_Sex.get(), entry_division.get(), entry_phone.get(),entry_email.get(), entry_address.get()))
    Staff_Table.insert(parent='',index='end',values=(entry_id.get(), entry_name.get(), combo_Sex.get(), entry_division.get(), entry_phone.get(), entry_email.get(), entry_address.get()))

def reset_info():
    entry_id.delete(0, len(entry_id.get()))
    entry_name.delete(0, len(entry_name.get()))
    entry_division.delete(0, len(entry_division.get()))
    entry_phone.delete(0, len(entry_phone.get()))
    entry_email.delete(0, len(entry_email.get()))
    entry_address.delete(0, len(entry_address.get()))
def delete_tree():
    selected_item = Staff_Table.selection()[0]
    Staff_Table.delete(selected_item)
def clear_all():
    for item in Staff_Table.get_children():
        Staff_Table.delete(item)
def delete_info():
    delete_id = entry_id.get()
    exist = False
    for item in range(len(list_staff)):
        if list_staff[item].id == entry_id.get():
            exist = True
            del list_staff[item]
            clear_all()
            for item in list_staff:
                Staff_Table.insert(parent='',index='end',values=(item.id, item.name, item.sex, item.division, item.phone, item.email, item.address))
            messagebox.showinfo("Xóa thông tin nhân viên", f"nhân viên có Id: {delete_id} đã được xoá")
            break
    if exist == False:
        messagebox.showinfo("Xoá Thông Tin nhân viên", f"nhân viên có id: {delete_id} không có")
def edit_info():
    check = entry_id.get()
    exist = False
    for item in range(len(list_staff)):
        if list_staff[item].id == check:
            list_staff[item].id = entry_id.get()
            list_staff[item].name = entry_name.get()
            list_staff[item].sex = combo_Sex.get()
            list_staff[item].division = entry_division.get()
            list_staff[item].phone = entry_phone.get()
            list_staff[item].email = entry_email.get()
            list_staff[item].address = entry_address.get()
            clear_all()
            for item in list_staff:
                Staff_Table.insert(parent='',index='end',values=(item.id, item.name, item.sex, item.division,item.phone, item.email, item.address))
            messagebox.showinfo("Sửa Thông Tin nhân viên","nhân viên co id:{check}  đã được sửa")
            exist = True
            break
    if exist == False:
        messagebox.showinfo("Sửa Thông Tin nhân viên ", f"nhân viên có id: {check} không tìm được trong danh sách")
def search_info():
    search = text_Search.get()
    exist = False
    if combo_Search.get() == "name":
        for item in list_staff:
            if item.name == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"nhân viên : {search} có tồn tại trong danh sách. id: {item.id}, name: {item.name}, sex: {item.sex}, division: {item.division}, phone: {item.phone}, email: {item.email}")
                break
    elif combo_Search.get() == "id":
        for item in list_staff:
            if item.id == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"id nhân viên: {search} có tồn tại trong danh sách. id: {item.id},name: {item.name}, sẽ: {item.sex}, division: {item.division}, phone: {item.phone},email: {item.email}")
                break
    elif combo_Search.get() == "division":
        for item in list_staff:
            if item.division == search:
                exist = True
                messagebox.showinfo("Tìm Kiếm", f"bộ phận: {search} có tồn tại trong danh sách. id: {item.id}, name:{item.name}, sex: {item.sex}, division: {item.division}, phone: {item.phone}, email: {item.mail}")
                break
    if exist == False:
        messagebox.showinfo("Tìm Kiếm", f"{search} không tồn tại trong danh sách")

root = Tk()
root.title("Staff Managerment System")
root.geometry('1000x700+0+0')
title = Label(root, text="Staff Managerment System", font=("arial",30,"bold"), fg="white", bg="green")
title.pack(side=TOP, fill=X)

Manager_Frame = Frame(root, bd=4, relief=RIDGE, bg="gray")
Manager_Frame.place(x=10, y=70, width=400, height=550)

m_title = Label(Manager_Frame, text="MANAGER STAFF", font=("arial",16,"bold"), bg="green", fg="white")
m_title.grid(row=0, columnspan=2, pady=10, padx=80)
        
lbl_id = Label(Manager_Frame, text=" ID", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_id.grid(row=1, column=0, pady=10, padx=20, sticky="w")     
entry_id = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_id.grid(row=1, column=1, pady=10, padx=20, sticky="w")
        
lbl_name = Label(Manager_Frame, text="Name", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_name.grid(row=2, column=0, pady=10, padx=20, sticky="w")
entry_name = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_name.grid(row=2, column=1, pady=10, padx=20, sticky="w")
        
lbl_sex = Label(Manager_Frame, text="Sex", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_sex.grid(row=3, column=0, pady=10, padx=20, sticky="w")
combo_Sex = ttk.Combobox(Manager_Frame, font=("arial",11), state="readonly")
combo_Sex['values']=("male","female","other")
combo_Sex.grid(row=3, column=1, ipady=1.5, ipadx=1)
        
lbl_division = Label(Manager_Frame, text="Division", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_division.grid(row=4, column=0, pady=10, padx=20, sticky="w")
entry_division = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_division.grid(row=4, column=1, pady=10, padx=20, sticky="w")
        
lbl_phone = Label(Manager_Frame, text="Phone", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_phone.grid(row=5, column=0, pady=10, padx=20, sticky="w")
entry_phone = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_phone.grid(row=5, column=1, pady=10, padx=20, sticky="w")

lbl_email = Label(Manager_Frame, text="Email", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_email.grid(row=6, column=0, pady=10, padx=20, sticky="w")
entry_email = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_email.grid(row=6, column=1, pady=10, padx=20, sticky="w")
        
lbl_address = Label(Manager_Frame, text="Address", font=("arial",12,"bold"), bg="gray", fg="white")
lbl_address.grid(row=7, column=0, pady=10, padx=20, sticky="w")
entry_address = Entry(Manager_Frame, font=("arial",12), bd=5, relief=GROOVE)
entry_address.grid(row=7, column=1, pady=10, padx=20, sticky="w")
        
#Button Frame
Btn_Frame = Frame(root, bd=4, relief=RIDGE, bg="white")
Btn_Frame.place(x=10, y=530, width=400)
        
Btn_Add = Button(Btn_Frame, text="Add", width=10, command=add_info)
Btn_Add.grid(row=0, column=0, pady=10, padx=10)
Btn_Reset = Button(Btn_Frame, text="Reset", width=10, command=reset_info)
Btn_Reset.grid(row=0, column=1, pady=10, padx=10)
Btn_Edit = Button(Btn_Frame, text="Edit", width=10, command=edit_info)
Btn_Edit.grid(row=0, column=2, pady=10, padx=10)
Btn_Delete = Button(Btn_Frame, text="Delete", width=10, command=delete_info)
Btn_Delete.grid(row=0, column=3, pady=10, padx=10)
        
# Manager detail       
Detail_Frame = Frame(root, bd=4, relief=RIDGE, bg="white")
Detail_Frame.place(x=420, y=70, width=850, height=550)
        
lbl_Search = Label(Detail_Frame, text="Search By", font=("arial",12,"bold"), bg="white", fg="black")
lbl_Search.grid(row=0, column=0, pady=10, padx=20, sticky="w")
        
combo_Search = ttk.Combobox(Detail_Frame, width=10, font=("arial",11), state="readonly")
combo_Search['values']=("id","name","division")
combo_Search.grid(row=0, column=1, ipady=1.5, ipadx=1)
        
text_Search = Entry(Detail_Frame, width=30, bd=5, relief=GROOVE)
text_Search.grid(row=0, column=2, pady=10, padx=20, sticky="w")
        
Btn_Search = Button(Detail_Frame, text="Search", width=10, pady=5, command=search_info)
Btn_Search.grid(row=0, column=3, pady=10, padx=10)
Btn_ShowAll = Button(Detail_Frame, text="Click & Delete", width=10, pady=5, command=delete_tree)
Btn_ShowAll.grid(row=0, column=4, pady=10, padx=10)
        
# Table Frame
Table_Frame = Frame(Detail_Frame, bd=4, relief=RIDGE, bg="white")
Table_Frame.place(x=10, y=70, width=820, height=460)
        
scroll_x = Scrollbar(Table_Frame, orient=HORIZONTAL)
scroll_y = Scrollbar(Table_Frame, orient=VERTICAL)
Staff_Table = ttk.Treeview(Table_Frame, columns=("id", "name", "sex", "division", "phone", "email", "address"), xscrollcommand=scroll_x.set, yscrollcommand=scroll_y.set)
scroll_x.pack(side=BOTTOM, fill=X)
scroll_y.pack(side=RIGHT, fill=Y)
scroll_x.config(command=Staff_Table.xview)
scroll_y.config(command=Staff_Table.yview)
        
Staff_Table.heading("id", text="ID")
Staff_Table.heading("name", text="Name")
Staff_Table.heading("sex", text="Sex")
Staff_Table.heading("division", text="Division")
Staff_Table.heading("phone", text="Phone")
Staff_Table.heading("email", text="Email")
Staff_Table.heading("address", text="Ddress")
        
Staff_Table['show'] = 'headings'
Staff_Table.column("id", width=100)
Staff_Table.column("name", width=100)
Staff_Table.column("sex", width=100)
Staff_Table.column("division", width=100)
Staff_Table.column("phone", width=100)
Staff_Table.column("email", width=100)
Staff_Table.column("address", width=200)
        
Staff_Table.pack(fill=BOTH, expand=1)

root.mainloop()
```
